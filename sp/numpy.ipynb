```python
#C1
def func_mean(x: list) -> float:
    return sum(x) / len(x)

#C2
def func_std(x: list, precomp_mean: float | None = None) -> tuple[float, float]:
    dof = len(x) - 1
    if precomp_mean is not None:
        mean = precomp_mean
    else:
        mean = func_mean(x)
    var = sum((val - mean) ** 2 for val in x) / dof
    return var, var ** 0.5

#C3
def func_zscore(x: list) -> list:
    mean = func_mean(x)
    _, std = func_std(x, precomp_mean=mean)
    return [(val - mean) / std for val in x]

#C4
def func_matrix_mean(M: list[list], axis: int = 0) -> list[list]:
    assert (axis == 0) or (axis == 1), "Axis must be 0 (col) or 1 (row)"
    if axis == 0:
        return [func_mean(row) for row in zip(*M)]
    elif axis == 1: 
        return [func_mean(row) for row in M]

#C5
def func_median(x: list) -> float:
    s = sorted(x)
    n = len(s)
    if n % 2 == 1:
        return s[n // 2]
    return 0.5 * (s[n // 2 - 1] + s[n // 2])

#C6
def func_corr(x: list, y: list):
    assert len(x) == len(y), "x, y must be the same length"
    dof = len(x) - 1
    mu_x, mu_y = func_mean(x), func_mean(y)
    (_, std_x), (__, std_y) = func_std(x, precomp_mean=mu_x), func_std(y, precomp_mean=mu_y)
    numerator = sum((xval - mu_x) * (yval - mu_y) for xval, yval in zip(x,y)) / dof
    if std_y * std_x != 0:
        return numerator / (std_x * std_y)
    else:
        raise ArithmeticError("Zero variance")

#C7
def func_matrix_cov(M: list[list], precomp_mean: list[list] | None = None):
    if precomp_mean is not None:
        mean = precomp_mean
    else:
        mean = func_matrix_mean(M=M, axis = 1)
    nrows, ncols = len(M), len(M-1)
    for rindex in range(nrows):
        chosen_row = M[rindex]
        for cindex in range(ncols):
            pass 
    return

#C8


#C9 
```
